```python

import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('iris')
df
```

```python

oh_species = pd.get_dummies(df['species'])
df = pd.concat([df.drop('species', axis=1), oh_species], axis=1)
```

```python
from scipy.stats import zscore
data=sns.load_dataset('iris')
df=data.copy()
```

```python
df
```

```python
(4.9-df['sepal_length'].mean())/df['sepal_length'].std()
```

```python
z_scores=np.abs(zscore(df.drop('species',axis=1)))
```

```python
(z_scores<=3).all(axis=1)
```

```python
outliers = (z_scores > 3).any(axis=1)
outlier_rows = df[outliers]
outlier_rows
```

```python
from sklearn.ensemble import IsolationForest
import numpy as np

data = np.array([[1, 2], [2, 3], [1, 3], [3, 4], [8, 7]])


clf = IsolationForest(random_state=10)

clf.fit(data)

predictions = clf.predict(data)

print(predictions)
```
